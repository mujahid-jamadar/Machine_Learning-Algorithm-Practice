```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('weight-height.csv',)
```

```python
df.head()
```

```python
# Check Null Values
df.isnull().sum()
```

```python
df.shape
```

### Lets check the correlation between each variables

```python
#Plot Gender vs Weight

plt.scatter(x=df.iloc[:,0],y=df.iloc[:,2])
plt.xlabel("Gender")
plt.ylabel('Weight')
plt.title('Gender vs Weight')
plt.legend()
plt.show()
```

### From above Plot we can see that weight is not all dependent on Gender

```python
#Plot Height vs Weight
plt.scatter(x=df.iloc[:,1],y=df.iloc[:,2])
plt.xlabel('Height')
plt.ylabel('Weight')
plt.title('Height Vs Weight')
plt.legend()
plt.show()
```

### From above plot we can see that  Weight and Height is highly dependend on each other.We can say that as height increases ,weight also increases.

```python
#separating the dependent and independent values

#X-Independent variable
x=df.iloc[:,1:2].values
x
```

```python
#Y-dependent variable
y=df.iloc[:,2].values
y
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
model.fit(x,y)
```

```python
# Check Accuracy
model.score(x,y)
```

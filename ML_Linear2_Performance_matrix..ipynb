# Splitting of dataset

```python
x=df.iloc[:,1:].values
```

```python
x[:2]
```

```python
y=df.mpg
```

```python
y[:2]
```

## Train Test Split

```python
from sklearn.model_selection import train_test_split
X_Train,X_Test,Y_Train,Y_Test=train_test_split(x,y,test_size=0.3)
```

```python
len(x)
```

```python
len(X_Train)
```

## lets build a model

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
model.fit(X_Train,Y_Train)
```

```python
#Testing
y_pred=model.predict(X_Test)
```

```python
y_pred
```

```python
Y_Test.values
```

```python
# PLot the graph of data
plt.scatter(X_Train[:,1],Y_Train)


# plot prediction line
plt.plot(X_Test[:,1],y_pred,'*r')
```

## How to measure the performance of Linear Regression/Multi Linear Regression model

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
# Actual Values
Y_Test[:2]
```

```python
# predicted Values
y_pred[:2]
```

```python
mean_absolute_error(Y_Test,y_pred)
```

```python
# MSE: squared error-> penalizes large error
mean_squared_error(Y_Test,y_pred)
```

```python
# r2_score:tell how best ur model is coef_determination
# value range 0-1
r2_score(Y_Test,y_pred)
```
